## 1. Setup and Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set visualization style
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
```

## 2. Understanding Histograms

### What is a Histogram?

A histogram is a visual representation of data distribution:
- **X-axis:** Range of values (divided into bins)
- **Y-axis:** Frequency (count of values in each bin)
- **Bars:** Height shows how many values fall in that range

### Histogram vs Bar Chart

**Histogram:**
- For continuous numeric data
- Bars touch each other
- Shows distribution shape

**Bar Chart:**
- For categorical data
- Bars are separated
- Shows comparisons between categories

### Why Histograms Matter

- Reveal patterns that summary statistics hide
- Show distribution shape (normal, skewed, bimodal)
- Identify outliers and unusual values visually
- Guide statistical analysis decisions

## 5. Creating a Single-Column Histogram

Let's create a histogram for the **Visitors** column.

```python
# Statistics for Visitors column
print("Statistics for Visitors:")
print(f"  Count:  {df['Visitors'].count()}")
print(f"  Mean:   {df['Visitors'].mean():.2f}")
print(f"  Median: {df['Visitors'].median():.2f}")
print(f"  Std:    {df['Visitors'].std():.2f}")
print(f"  Min:    {df['Visitors'].min():.2f}")
print(f"  Max:    {df['Visitors'].max():.2f}")
```

```python
# Create histogram for Visitors
plt.figure(figsize=(10, 6))
plt.hist(df['Visitors'].dropna(), bins=10, color='steelblue', 
         edgecolor='black', alpha=0.7)
plt.xlabel('Visitors', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Visitors', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Interpretation

**What to look for:**
- Where are most values concentrated?
- Is the distribution symmetric or skewed?
- Are there any gaps or unusual patterns?
- Do you see potential outliers?

## 6. Interpreting Distribution Shape

```python
# Calculate distribution characteristics
visitors_data = df['Visitors'].dropna()
mean_val = visitors_data.mean()
median_val = visitors_data.median()
skewness = visitors_data.skew()
std_val = visitors_data.std()

print("Distribution Characteristics for Visitors:")
print(f"  Mean:     {mean_val:.2f}")
print(f"  Median:   {median_val:.2f}")
print(f"  Skewness: {skewness:.3f}")
print(f"  Std Dev:  {std_val:.2f}")
print()

# Interpret skewness
if abs(skewness) < 0.5:
    print("✓ Approximately Symmetric (Normal-like)")
    print("  → Mean ≈ Median")
    print("  → Data is evenly distributed around the center")
elif skewness > 0.5:
    print("✓ Right-Skewed (Positively Skewed)")
    print("  → Mean > Median")
    print("  → Long tail extends to the right")
    print("  → Most values are on the lower end")
else:
    print("✓ Left-Skewed (Negatively Skewed)")
    print("  → Mean < Median")
    print("  → Long tail extends to the left")
    print("  → Most values are on the higher end")
```

## 7. Histogram for Revenue

```python
# Create histogram for Revenue
plt.figure(figsize=(10, 6))
plt.hist(df['Revenue'].dropna(), bins=10, color='forestgreen', 
         edgecolor='black', alpha=0.7)
plt.xlabel('Revenue', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Revenue', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Statistics
print("\nStatistics for Revenue:")
print(f"  Mean:     {df['Revenue'].mean():.2f}")
print(f"  Median:   {df['Revenue'].median():.2f}")
print(f"  Skewness: {df['Revenue'].skew():.3f}")
```

## 8. Comparing Multiple Histograms

```python
# Create side-by-side histograms
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Visitors
axes[0].hist(df['Visitors'].dropna(), bins=10, color='steelblue', 
             edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Visitors', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].set_title('Visitors Distribution', fontsize=12, fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)
axes[0].axvline(df['Visitors'].mean(), color='red', linestyle='--', 
                linewidth=2, label=f"Mean: {df['Visitors'].mean():.1f}")
axes[0].legend()

# Revenue
axes[1].hist(df['Revenue'].dropna(), bins=10, color='forestgreen', 
             edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Revenue', fontsize=11)
axes[1].set_ylabel('Frequency', fontsize=11)
axes[1].set_title('Revenue Distribution', fontsize=12, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)
axes[1].axvline(df['Revenue'].mean(), color='red', linestyle='--', 
                linewidth=2, label=f"Mean: {df['Revenue'].mean():.1f}")
axes[1].legend()

# Rating
axes[2].hist(df['Rating'].dropna(), bins=10, color='coral', 
             edgecolor='black', alpha=0.7)
axes[2].set_xlabel('Rating', fontsize=11)
axes[2].set_ylabel('Frequency', fontsize=11)
axes[2].set_title('Rating Distribution', fontsize=12, fontweight='bold')
axes[2].grid(axis='y', alpha=0.3)
axes[2].axvline(df['Rating'].mean(), color='red', linestyle='--', 
                linewidth=2, label=f"Mean: {df['Rating'].mean():.1f}")
axes[2].legend()

plt.tight_layout()
plt.show()
```

### Comparison Summary

```python
# Create comparison table
comparison = pd.DataFrame({
    'Column': ['Visitors', 'Revenue', 'Rating'],
    'Mean': [
        df['Visitors'].mean(),
        df['Revenue'].mean(),
        df['Rating'].mean()
    ],
    'Median': [
        df['Visitors'].median(),
        df['Revenue'].median(),
        df['Rating'].median()
    ],
    'Skewness': [
        df['Visitors'].skew(),
        df['Revenue'].skew(),
        df['Rating'].skew()
    ],
    'Std Dev': [
        df['Visitors'].std(),
        df['Revenue'].std(),
        df['Rating'].std()
    ]
})

print("Distribution Comparison:")
print(comparison.to_string(index=False))
```

## 9. Effect of Bin Size

```python
# Show different bin sizes for the same data
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

bin_sizes = [5, 10, 20]

for idx, bins in enumerate(bin_sizes):
    axes[idx].hist(df['Visitors'].dropna(), bins=bins, color='steelblue', 
                   edgecolor='black', alpha=0.7)
    axes[idx].set_xlabel('Visitors', fontsize=11)
    axes[idx].set_ylabel('Frequency', fontsize=11)
    axes[idx].set_title(f'{bins} Bins', fontsize=12, fontweight='bold')
    axes[idx].grid(axis='y', alpha=0.3)

plt.suptitle('Effect of Bin Size on Visitors Distribution', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("\nBin Size Guidelines:")
print("  5 bins:  Simple, smooth view - good for overview")
print("  10 bins: Balanced detail - default choice")
print("  20 bins: Detailed view - reveals fine patterns")
```

## 10. Detecting Outliers Visually

```python
# Outlier detection using IQR method
Q1 = df['Visitors'].quantile(0.25)
Q3 = df['Visitors'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df['Visitors'][(df['Visitors'] < lower_bound) | (df['Visitors'] > upper_bound)]

print("Outlier Analysis for Visitors:")
print(f"  Q1: {Q1:.2f}")
print(f"  Q3: {Q3:.2f}")
print(f"  IQR: {IQR:.2f}")
print(f"  Lower Bound: {lower_bound:.2f}")
print(f"  Upper Bound: {upper_bound:.2f}")
print(f"\n  Outliers Detected: {len(outliers)}")
if len(outliers) > 0:
    print(f"  Outlier Values: {outliers.values}")
```
