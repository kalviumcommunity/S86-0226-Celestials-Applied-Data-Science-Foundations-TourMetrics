## 1. Understanding CSV Files

### What is a CSV File?

CSV stands for **Comma-Separated Values**. It's a simple text format for storing tabular data:
- Each line represents a **row**
- Values are separated by commas (the **delimiter**)
- The first row usually contains **column headers**
- Think of it like a spreadsheet saved as plain text

### Why Use Pandas for CSV Loading?

Pandas makes CSV loading easy:
- One simple function: `pd.read_csv()`
- Automatically detects data types
- Handles headers intelligently
- Provides powerful inspection tools
- Industry standard for data analysis

````python
# Import the libraries we need
import pandas as pd
import os

print("✓ Libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
````

## 2. Loading Your First CSV File

Let's load the `books.csv` file. The `pd.read_csv()` function does all the work:
- Reads the file from the specified path
- Parses the CSV format
- Creates a DataFrame object
- Returns the DataFrame for you to use

````python
# Load the books.csv file
books_df = pd.read_csv('data/raw/books.csv')

print("✓ Successfully loaded books.csv!")
print(f"Created a DataFrame named 'books_df'")
````

## 3. Inspecting Loaded Data

**Always inspect your data after loading!** This prevents silent errors and ensures your analysis starts correctly.

### Method 1: head() - Preview the First Rows

The `head()` method shows the first 5 rows by default. This gives you a quick look at your data structure.

````python
# Display the first 5 rows
books_df.head()
````

**Expected Output:**
```

0                   1984        George Orwell  1949    328
1  To Kill a Mockingbird           Harper Lee  1960    281
2       The Great Gatsby  F. Scott Fitzgerald  1925    180
3    Pride and Prejudice          Jane Austen  1813    432
```

**What to notice:**
- Column names are at the top: Title, Author, Year, Pages
- Row numbers (index) start at 0
- Data is neatly organized in a table format

### Method 2: tail() - Preview the Last Rows

The `tail()` method shows the last 5 rows. This helps verify the file loaded completely.

````python
# Display the last 5 rows
books_df.tail()
````

### Method 3: info() - Get DataFrame Information

The `info()` method provides a comprehensive summary:
- Number of rows and columns
- Column names
- Data types for each column
- Count of non-null values (helps spot missing data)
- Memory usage

````python
# Get detailed information about the DataFrame
books_df.info()
````

**Expected Output:**
```
<class 'pandas.DataFrame'>
RangeIndex: 4 entries, 0 to 3
Data columns (total 4 columns):
 #   Column  Non-Null Count  Dtype
---  ------  --------------  -----
 0   Title   4 non-null      str
 1   Author  4 non-null      str
 2   Year    4 non-null      int64
 3   Pages   4 non-null      int64
```

**Key insights:**
- 4 rows (entries 0 to 3)
- 4 columns total
- No missing values (all show "4 non-null")
- Title and Author are strings (str)
- Year and Pages are integers (int64)

### Method 4: describe() - Statistical Summary

The `describe()` method provides statistics for numeric columns:
- count: Number of values
- mean: Average value
- std: Standard deviation
- min/max: Minimum and maximum values
- 25%, 50%, 75%: Quartiles

````python
# Get statistical summary of numeric columns
books_df.describe()
````

**Expected Output:**
```

std      67.435772  104.656183
min    1813.000000  180.000000
25%    1897.000000  255.750000
50%    1937.000000  304.500000
75%    1951.750000  354.000000
max    1960.000000  432.000000
```

**What this tells us:**
- Books range from 1813 to 1960
- Average book has about 305 pages
- Page counts range from 180 to 432

## 6. Common CSV Loading Issues

Understanding what can go wrong helps you troubleshoot problems quickly.

### Issue 1: File Not Found

**Problem:** The file path is incorrect or the file doesn't exist.

**Solution:** Check the file path and verify the file exists.

````python
# Example: Handling file not found
try:
    df = pd.read_csv('data/raw/nonexistent.csv')
except FileNotFoundError as e:
    print("✗ Error: File not found")
    print(f"Details: {e}")
    print("\nTip: Check that the file path is correct")
````

````python
# Better approach: Check if file exists first
filepath = 'data/raw/books.csv'

if os.path.exists(filepath):
    df = pd.read_csv(filepath)
    print(f"✓ Successfully loaded {filepath}")
else:
    print(f"✗ File not found: {filepath}")
````

### Issue 2: Encoding Problems

**Problem:** File uses a different character encoding (not UTF-8).

**Solution:** Specify the encoding parameter.

````python
# If you get UnicodeDecodeError, try different encodings:

# Option 1: Latin-1 encoding
# df = pd.read_csv('file.csv', encoding='latin-1')

# Option 2: Windows encoding
# df = pd.read_csv('file.csv', encoding='cp1252')

# Option 3: Explicitly specify UTF-8
# df = pd.read_csv('file.csv', encoding='utf-8')

print("Encoding examples shown above (commented out)")
````

### Issue 3: Wrong Delimiter

**Problem:** File uses semicolons, tabs, or other delimiters instead of commas.

**Solution:** Use the `sep` parameter to specify the delimiter.

````python
# For semicolon-separated files:
# df = pd.read_csv('file.csv', sep=';')

# For tab-separated files:
# df = pd.read_csv('file.tsv', sep='\t')

# For pipe-separated files:
# df = pd.read_csv('file.txt', sep='|')

print("Delimiter examples shown above (commented out)")
````

**Problem:** File has no header row, or header is on a different line.

**Solution:** Use the `header` parameter.
