# Pandas DataFrame Fundamentals

This notebook covers:
- Understanding DataFrames
- Creating DataFrames from dictionaries
- Loading DataFrames from files
- Inspecting DataFrame structure

```python
import pandas as pd
import numpy as np
```

## 1. Understanding Pandas DataFrames

A DataFrame is a 2-dimensional labeled data structure with columns of potentially different types.
Think of it as:
- A spreadsheet or SQL table
- A dictionary of Series objects
- Rows and columns with labels

## 2. Creating DataFrames from Dictionaries

The most common way to create a DataFrame programmatically is from a dictionary.

```python
# Create a dictionary with column-style data
student_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [23, 25, 22, 24],
    'Grade': ['A', 'B', 'A', 'B'],
    'Score': [92, 85, 95, 88]
}

# Convert dictionary to DataFrame
df_students = pd.DataFrame(student_data)
print("DataFrame from Dictionary:")
print(df_students)
```

```python
# Inspect column names
print("\nColumn Names:")
print(df_students.columns.tolist())
```

```python
# Inspect index
print("\nIndex:")
print(df_students.index.tolist())
```

```python
# Another example - sales data
sales_data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [999.99, 25.50, 75.00, 299.99],
    'Quantity': [5, 50, 30, 10]
}

df_sales = pd.DataFrame(sales_data)
print("\nSales DataFrame:")
print(df_sales)
```

## 3. Creating DataFrames from Files

Loading data from external files is the most common real-world scenario.

```python
# First, let's create a sample CSV file for demonstration
sample_csv_data = """Name,Department,Salary,Years
John,Engineering,75000,3
Sarah,Marketing,65000,2
Mike,Engineering,80000,5
Emma,Sales,70000,4
David,Marketing,68000,3"""

# Save to file
with open('../data/raw/employees.csv', 'w') as f:
    f.write(sample_csv_data)

print("Sample CSV file created!")
```

```python
# Load DataFrame from CSV file
df_employees = pd.read_csv('../data/raw/employees.csv')
print("DataFrame loaded from CSV:")
print(df_employees)
```

```python
# Understanding how headers and rows are interpreted
print("\nHeaders (Column Names):")
print(df_employees.columns.tolist())
print("\nFirst row of data:")
print(df_employees.iloc[0])
```

## 4. Inspecting DataFrame Structure

Always inspect your data after loading to catch issues early.

```python
# View first few rows
print("First 3 rows:")
print(df_employees.head(3))
```

```python
# View last few rows
print("\nLast 2 rows:")
print(df_employees.tail(2))
```

```python
# Check shape (rows, columns)
print("\nDataFrame Shape:")
print(f"Rows: {df_employees.shape[0]}, Columns: {df_employees.shape[1]}")
```

```python
# Inspect column names
print("\nColumn Names:")
print(df_employees.columns.tolist())
```

```python
# Check data types
print("\nData Types:")
print(df_employees.dtypes)
```

```python
# Get comprehensive info
print("\nDataFrame Info:")
df_employees.info()
```

```python
# Basic statistics
print("\nBasic Statistics:")
print(df_employees.describe())
```

## Practice Examples

Let's practice with different scenarios.

```python
# Example 1: Weather data
weather_data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Temperature': [72, 85, 65, 90],
    'Humidity': [60, 45, 70, 75],
    'Condition': ['Sunny', 'Clear', 'Cloudy', 'Hot']
}

df_weather = pd.DataFrame(weather_data)
print("Weather DataFrame:")
print(df_weather)
print(f"\nShape: {df_weather.shape}")
print(f"Columns: {df_weather.columns.tolist()}")
```

```python
# Example 2: Create and save another CSV
books_csv = """Title,Author,Year,Pages
1984,George Orwell,1949,328
To Kill a Mockingbird,Harper Lee,1960,281
The Great Gatsby,F. Scott Fitzgerald,1925,180
Pride and Prejudice,Jane Austen,1813,432"""

with open('../data/raw/books.csv', 'w') as f:
    f.write(books_csv)

# Load and inspect
df_books = pd.read_csv('../data/raw/books.csv')
print("\nBooks DataFrame:")
print(df_books.head())
print("\nData Types:")
print(df_books.dtypes)
```

## Key Takeaways

1. **DataFrames are 2D structures** - rows and columns with labels
2. **Dictionaries to DataFrames** - keys become column names, values become data
3. **File loading** - `pd.read_csv()` is the most common method
4. **Always inspect** - use `.head()`, `.shape`, `.dtypes`, `.info()`
5. **DataFrames are the foundation** - all Pandas analysis starts here
