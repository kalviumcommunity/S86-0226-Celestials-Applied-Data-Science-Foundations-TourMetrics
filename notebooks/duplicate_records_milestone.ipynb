# Duplicate Records Detection and Removal Milestone

This notebook demonstrates how to identify and remove duplicate records in Pandas DataFrames.

## Learning Objectives
- Understand what duplicate records are
- Detect duplicate rows in a DataFrame
- Remove duplicates safely and intentionally
- Verify results after deduplication

```python
import pandas as pd
import numpy as np
```

## 1. Understanding Duplicate Records

Duplicate records are rows that have identical values across all or selected columns. They can occur due to:
- Data entry errors
- System glitches
- Merging datasets
- Import/export issues

```python
# Load dataset with duplicates
df = pd.read_csv('../data/raw/products.csv')

print("Original Dataset:")
print(df)
print(f"\nDataset shape: {df.shape}")
print(f"Total records: {len(df)}")
```

## 2. Detecting Duplicate Rows

Use `duplicated()` to identify duplicate rows. It returns a boolean Series.

```python
# Check for duplicates
duplicates_mask = df.duplicated()

print("Boolean mask showing duplicate rows:")
print(duplicates_mask)
print(f"\nNumber of duplicates: {duplicates_mask.sum()}")
print(f"Percentage: {(duplicates_mask.sum() / len(df)) * 100:.2f}%")
```

```python
# Show duplicate rows
print("Rows marked as duplicates:")
df[duplicates_mask]
```

```python
# Show all occurrences of duplicated records (including first)
duplicates_all = df.duplicated(keep=False)

print("All occurrences of duplicated records:")
df[duplicates_all].sort_values('ProductID')
```

## 3. Detecting Duplicates on Specific Columns

Sometimes you only want to check for duplicates based on certain columns.

```python
# Check duplicates based on ProductID only
duplicates_by_id = df.duplicated(subset=['ProductID'])

print("Duplicates based on ProductID:")
print(df[duplicates_by_id])
print(f"\nCount: {duplicates_by_id.sum()}")
```

```python
# Check duplicates based on multiple columns
duplicates_by_name = df.duplicated(subset=['ProductName', 'Category'])

print("Duplicates based on ProductName and Category:")
print(df[duplicates_by_name])
print(f"\nCount: {duplicates_by_name.sum()}")
```

## 4. Removing Duplicate Records

Use `drop_duplicates()` to remove duplicate rows.

```python
print("Before deduplication:")
print(f"Shape: {df.shape}")
print(f"Total records: {len(df)}")
```

```python
# Remove duplicates (keeps first occurrence by default)
df_cleaned = df.drop_duplicates()

print("After deduplication:")
print(df_cleaned)
print(f"\nShape: {df_cleaned.shape}")
print(f"Records removed: {len(df) - len(df_cleaned)}")
```

```python
# Remove duplicates keeping last occurrence
df_cleaned_last = df.drop_duplicates(keep='last')

print("Keeping last occurrence:")
print(f"Shape: {df_cleaned_last.shape}")
```

```python
# Remove duplicates based on specific columns
df_cleaned_by_id = df.drop_duplicates(subset=['ProductID'])

print("Deduplication based on ProductID only:")
print(df_cleaned_by_id)
print(f"\nShape: {df_cleaned_by_id.shape}")
```

## 5. Verifying Deduplication Results

Always verify that deduplication worked correctly.

```python
print("Verification Summary:")
print("-" * 50)
print(f"Original dataset size: {len(df)} rows")
print(f"Cleaned dataset size: {len(df_cleaned)} rows")
print(f"Duplicates removed: {len(df) - len(df_cleaned)} rows")
print(f"Reduction: {((len(df) - len(df_cleaned)) / len(df)) * 100:.2f}%")
```

```python
# Verify no duplicates remain
remaining_duplicates = df_cleaned.duplicated().sum()
print(f"Remaining duplicates: {remaining_duplicates}")

if remaining_duplicates == 0:
    print("✓ Deduplication successful!")
else:
    print("⚠ Warning: Duplicates still present")
```

```python
# Compare unique values
print("Unique ProductIDs:")
print(f"Original: {df['ProductID'].nunique()}")
print(f"Cleaned: {df_cleaned['ProductID'].nunique()}")
```

## 6. Saving Cleaned Data

```python
# Save cleaned data
output_path = '../data/processed/products_cleaned.csv'
df_cleaned.to_csv(output_path, index=False)
print(f"Cleaned data saved to: {output_path}")
```

## Key Takeaways

- `duplicated()` identifies duplicate rows (returns boolean mask)
- `drop_duplicates()` removes duplicate rows
- `keep` parameter controls which duplicate to retain ('first', 'last', False)
- `subset` parameter allows checking specific columns
- Always verify results after deduplication
- Deduplication improves data quality and analysis accuracy
