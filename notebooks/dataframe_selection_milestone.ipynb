# DataFrame Selection Milestone - Indexing and Slicing

**Learning Objectives:**
- Select columns by name (single and multiple)
- Select rows by position using integer-based indexing
- Select rows by label using label-based indexing
- Combine row and column selection safely
- Understand when to use each selection method

---

````python
# Import required libraries
import pandas as pd
import numpy as np

# Display settings for better output
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
````

## 1. Creating Sample Data

Let's create a Tour Performance dataset to practice selection techniques.

````python
# Create sample Tour Performance data
tour_data = {
    'TourID': ['T001', 'T002', 'T003', 'T004', 'T005', 'T006', 'T007', 'T008'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 
             'Philadelphia', 'San Antonio', 'San Diego'],
    'Attendance': [15000, 18500, 12000, 14200, 11000, 13500, 9800, 16700],
    'Revenue': [450000, 555000, 360000, 426000, 330000, 405000, 294000, 501000],
    'Rating': [4.5, 4.8, 4.2, 4.6, 4.0, 4.4, 3.9, 4.7],
    'Duration_Hours': [2.5, 3.0, 2.0, 2.5, 2.0, 2.5, 2.0, 3.0]
}

df_tours = pd.DataFrame(tour_data)
print("Sample Tour Performance Dataset:")
display(df_tours)
print(f"\nShape: {df_tours.shape}")
print(f"Columns: {df_tours.columns.tolist()}")
````

## 2. Selecting Columns by Name

**Key Points:**
- Single brackets + string → Series (1D)
- Double brackets + list → DataFrame (2D)
- Column order in list determines output order

````python
# Single column selection (returns Series)
print("Single Column Selection:")
city_series = df_tours['City']
print(city_series)
print(f"\nType: {type(city_series)}")
````

````python
# Multiple column selection (returns DataFrame)
print("Multiple Column Selection:")
selected_columns = df_tours[['City', 'Attendance', 'Revenue']]
display(selected_columns)
print(f"\nType: {type(selected_columns)}")
````

````python
# Reordering columns
print("Reordering Columns:")
reordered = df_tours[['Revenue', 'Attendance', 'City']]
display(reordered)
````

## 3. Selecting Rows by Position (.iloc)

**Key Points:**
- `.iloc` uses integer positions (0-based)
- Slicing is EXCLUSIVE of end index
- Negative indexing works (e.g., -1 is last row)

````python
# Single row selection
print("First Row:")
first_row = df_tours.iloc[0]
print(first_row)
print(f"\nType: {type(first_row)}")
````

````python
# Multiple specific rows
print("Specific Rows (0, 2, 5):")
specific_rows = df_tours.iloc[[0, 2, 5]]
display(specific_rows)
````

````python
# Slicing rows (EXCLUSIVE of end index)
print("Rows 1 to 3 (excludes 4):")
sliced_rows = df_tours.iloc[1:4]
display(sliced_rows)
````

````python
# First n rows
print("First 3 Rows:")
display(df_tours.iloc[:3])
````

````python
# Last n rows
print("Last 3 Rows:")
display(df_tours.iloc[-3:])
````

````python
# Every nth row
print("Every 2nd Row:")
display(df_tours.iloc[::2])
````

## 4. Selecting Rows by Label (.loc)

**Key Points:**
- `.loc` uses index labels (not positions)
- Slicing is INCLUSIVE of both start and end
- Works with any index type (strings, dates, numeric)

````python
# First, create DataFrame with custom index
df_labeled = df_tours.set_index('TourID')
print("DataFrame with Custom Index:")
display(df_labeled)
````

````python
# Single row by label
print("Row with TourID 'T003':")
row_t003 = df_labeled.loc['T003']
print(row_t003)
````

````python
# Multiple specific rows by label
print("Rows T001, T004, T007:")
specific_rows = df_labeled.loc[['T001', 'T004', 'T007']]
display(specific_rows)
````

````python
# Slicing by label (INCLUSIVE of both ends!)
print("Rows T002 to T005 (INCLUSIVE):")
sliced_rows = df_labeled.loc['T002':'T005']
display(sliced_rows)
print("\n⚠️ Notice: Both T002 AND T005 are included!")
````

## 5. Combining Row and Column Selection

**Format:** `df.iloc[rows, columns]` or `df.loc[rows, columns]`

Use `:` to select all rows or all columns

````python
# Position-based: specific rows and columns
print("First 3 rows, columns at positions 1-3:")
subset1 = df_tours.iloc[0:3, 1:4]
display(subset1)
````

````python
# Position-based: specific rows, all columns
print("First 3 rows, all columns:")
subset2 = df_tours.iloc[0:3, :]
display(subset2)
````

````python
# Position-based: all rows, specific columns
print("All rows, columns at positions 1, 2, 4:")
subset3 = df_tours.iloc[:, [1, 2, 4]]
display(subset3)
````

````python
# Label-based: specific rows and columns
print("Rows T002 to T004, specific columns:")
subset4 = df_labeled.loc['T002':'T004', ['City', 'Attendance', 'Rating']]
display(subset4)
````

````python
# Extracting single value
print("Single Value Extraction:")
print("\nUsing .iloc (row 2, column 3):")
value1 = df_tours.iloc[2, 3]
print(f"Value: {value1}, Type: {type(value1)}")

print("\nUsing .loc (TourID 'T005', column 'Revenue'):")
value2 = df_labeled.loc['T005', 'Revenue']
print(f"Value: {value2}, Type: {type(value2)}")
````

## 6. Practical Examples

Apply selection techniques to real-world scenarios

````python
# Example 1: Get top 3 cities by revenue
print("Top 3 Cities by Revenue:")
df_sorted = df_tours.sort_values('Revenue', ascending=False)
top_performers = df_sorted[['City', 'Revenue', 'Rating']].iloc[:3]
display(top_performers)
````

````python
# Example 2: Compare middle tours
print("Middle Tours (indices 2-5):")
middle_tours = df_tours[['City', 'Attendance', 'Revenue']].iloc[2:6]
display(middle_tours)
````

````python
# Example 3: Revenue for specific cities
print("Revenue for Specific Cities:")
cities_of_interest = ['New York', 'Los Angeles', 'Chicago']
revenue_check = df_tours.loc[df_tours['City'].isin(cities_of_interest), ['City', 'Revenue']]
display(revenue_check)
````

## Quick Reference Guide

```
COLUMNS
df['A']              → Single column (Series)
df[['A', 'B']]       → Multiple columns (DataFrame)

ROWS BY POSITION (.iloc)
df.iloc[0]           → First row (Series)
df.iloc[0:3]         → Rows 0, 1, 2 (excludes 3)
df.iloc[[0, 2, 5]]   → Specific rows
df.iloc[-1]          → Last row

ROWS BY LABEL (.loc)
df.loc[0]            → Row with index label 0
df.loc[0:3]          → Rows 0, 1, 2, 3 (INCLUSIVE!)
df.loc[['A', 'B']]   → Rows with labels 'A' and 'B'

COMBINED SELECTION
df.iloc[0:3, 1:4]    → Rows 0-2, Columns 1-3
df.loc[:, ['A', 'B']]→ All rows, Columns 'A' and 'B'
df.iloc[2, 3]        → Single value at row 2, column 3

KEY DIFFERENCE
.iloc → Position-based, EXCLUSIVE slicing
.loc  → Label-based, INCLUSIVE slicing
```
